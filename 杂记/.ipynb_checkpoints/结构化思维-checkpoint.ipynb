# 什么是结构化思维
    结构化思维的定义就是: 逻辑 + 套路
## 表达要有逻辑
所谓逻辑是指我们的结构之间必须是有逻辑关系的.四种组织思想的逻辑关系: 
1. 演绎(因果)顺序:"大前提/小前提/结论"的演绎推理方式就是演绎顺序.比如,经典三段论:所有人都要死,苏格拉底是人,苏格拉底要死.
2. 时间(步骤)顺序:"第一/第二/第三","首先/然后/再者"等.
3. 空间(结构)顺序:"前端/后端/数据",化整为零等.
4. 程度(重要性)顺序:"最重要/次重要/不重要"

## 做事要有套路
套路是指我们解决问题的方法论 --> 5H2H 分析法,从 Why/Who/When/Where/What,How和How much 七个方面去思考.

# 如何进行结构化思考
进行结构化思考的两个步骤: 首先是"建立中心",然后再进行"分解".
## 建立中心
建立中心也就是要定义清楚要解决的问题,要明确目标. -- 先搞清楚 why,然后进行how.
建立中心有两种方式:
1. 自上而下:适用于问题比较明确的情况,只需找到问题的核心要素,然后进行展开
2. 自下而上:对于问题不够明确的情况,对多种杂乱的内容进行分类/剪枝/归纳汇总成一个中心

## 结构化分解
分解的策略即是四种逻辑顺序:演绎顺序/时间顺序/空间顺序/程度顺序.
空间分解 -- 注意满足 MECE(Mutually Exclusive Collectively Exhaustive,相互独立,完全穷尽)原则.

# 结构化思维应用
## 如何落地新团队?
1. 熟悉业务
    1. 了解产品
    2. 了解流程
    3. 客户走访
2. 熟悉技术
    1. 了解系统架构
    2. 了解领域模型
    3. 了解代码结构
3. 熟悉人
    1. 了解组织结构
    2. 了解人员角色
    3. 拜山头

##   打造极客文化
1. 空间顺序
    1. 技术沉淀
        1. 建立团队技术圈子
        2. 技术协会圈子运营
    2. 技术分享
    3. 技术氛围
        1. CR周报(代码好坏味道)
        2. 工匠月报(正向激励)
        3. Geekthon代码大赛
    4. 技术能力
        1. 架构委员会
        2. 技术委员会
2. 时间顺序
    1. 建立技术委员会
    2. 分配任务,指定owner
    3. 全员KO
    4. 安排执行
    5. 回顾复盘

## 如何做晋升述职
1. 罗列事情
    结构化的表达: 提出问题,定义问题,分析问题,解决问题,展望未来
    why -- 把价值说清楚 ; how -- 把过程讲明白
2. 价值的背后
    主导的风控系统将公司坏账率从5%降到2%,有表述还不够,还需要把价值背后的过程和思考说清楚:
    1. 之前为什么那么高?
    2. 为什么你的方法可以降低?是如何归因的?
    3. 具体解决了什么问题?
    4. 是否可以总结出一套方法,以后别人也能用这个方法解决这些问题?
