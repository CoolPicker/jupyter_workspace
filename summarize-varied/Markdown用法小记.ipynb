# Markdown用法整理
Markdown是一种轻量级标记语言，2004年由John Gruber创建。<br>
Markdown编写的文档可以到处HTML、Word、图像、PDF、Epub等多种格式的文档。<br>
Markdown编写的文档后缀为.md，.markdown。

## Markdown标题
```
使用#号标记：
使用#可以标记1-6级标题
# 一级标题
## 二级标题
### 三级标题
#### 四级标题
##### 五级标题
###### 六级标题
```
```
使用-和=标记语法格式
一级标题
---------------
二级标题
================
```

## Markdown段落

### 分段书写
我爱你中国  
亲爱的母亲  
我为你流泪  
也为你自豪  

Markdown段落没有特殊的格式，直接编写文字就好，**段落的换行是使用两个以上空格加上回车**。  
![段落示例1](https://www.runoob.com/wp-content/uploads/2019/03/3F254936-778E-417A-BEF2-467116A55D00.jpg)
当然也可以在段落后面使用一个空行来表示重新开始一个段落。
![段落实例2](https://www.runoob.com/wp-content/uploads/2019/03/3F254936-778E-417A-BEF2-467116A55D00.jpg)
### 字体(注意 * _ 均为英文状态下字符)
*斜体文本*  
_斜体文本_  
**粗体文本**  
__粗体文本__  
***粗斜体文本***  
___粗斜体文本___  
### 分割线
可以在一行中用三个以上的 * - _ 来建立分割线，行内不能有其他东西。也可以在其中加空格的方式。  
**注意： - 不推荐使用，容易与一级标题冲突 **
分割线  

***
分割线  

* * *
分割线一  

---
分割线一  

- - - 
分割线一  
___
分割线一  
_ _ _

### 删除线（两个波浪线 ~~ ）
~~知道错了~~
### 下划线（使用HTML的 &lt;u&gt; 标签）
<u>下划线</u>
### 脚注
脚注是对文本的补充说明  
`
[^要注明的文本]
`  

[^林则徐]离我越来越远啦！  

[^林则徐]: 苟利国家生死以，岂因祸福避趋之。

## Markdown列表
Markdown支持有序列表和无序列表，二者可自由嵌套，子列表前添加制表符 /tab 即可。  
**无序列表使用 * + - 作为列表标记：**
* first
* second
* third
+ first
+ second
+ third
- first
- second
- third

**有序列表使用数字加 . 表示：**  
1. first item
2. second item
3. third item
    1. test first
    2. test second
    3. test third
        1. go first 
        2. go second 
        3. go third
            - love one
            - love two
            - love three

## Markdown区块
Markdown区块引用是在段落开头使用 &gt; 符号，然后紧跟一个 **空格** 符号：
> 赵客缦胡缨，污垢双雪梅。  
> 银鞍照白马，飒沓如流星。  
> 十步杀一人，千里不留行。  
> 事了拂衣去，深藏功与名。  

另外区块可以嵌套，并可组合列表使用
> block first  
>> first go go  
>> first do do  
>> - 111  
>> - 222  
>> - 333  
> block second  
> block third  

- 11
> love  
> bee  
- 22

## Markdown链接
链接使用方法如下：  

    [链接名称](链接地址)
    或者
    <链接地址>
    
这是一个链接 [Google](https://www.google.com/)  
<https://www.google.com/>  
高级链接

    链接也可以用变量来代替，文档末尾附带变量地址：
    这个链接用 1 作为网址变量 [Google][1]
    这个链接用 Alphabet 作为网址变量 [Google][Alphabet]
    然后在文档的结尾为变量赋值(网址)。
    
    [1]: https://www.google.com/
    [Alphabet]: https://www.google.com/

## Markdown 表格
Markdown制作表格使用 \| 来分割不同的单元格，使用 \- 来分割表头和其他行。  
语法格式如下：

    | 姓名 | 年龄 | 
    | --- | --- |
    | 小明 | 28 |
    | 小芳 | 16 |
    
| 姓名 | 年龄 | 
| --- | --- |
| 小明 | 28 |
| 小芳 | 16 |

**对齐方式：**  
+ `-:`  居右对齐
+ `:-`  居左对齐
+ `:-:` 居中对齐

| 姓　　名 | 年　　龄 | 性　　别 |
| :--- | ---: | :---: |
| 小明 | 28 | 男 |
| 小芳 | 16 | 女 |
