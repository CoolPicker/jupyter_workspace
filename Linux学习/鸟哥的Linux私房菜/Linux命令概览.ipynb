## Linux的文件权限和目录配置

改变文件属性和权限:  
* chgrp: 改变文件所属群组
* chown: 改变文件拥有者
* chmod: 改变文件的权限

```bash
    chgrp [-R] user dir/file
    chown [-R] user dir/file
    chmod [-R] 137 dir/file
# 1 2 4 3 5 6 7
# Linux文件基本权限,九个,分别是owner/group/others
# 三种身份对应各自的read/write/execute权限
```  
文件和和文件夹权限汇总:  
|元件|内容|叠代物件|r|w|x|
|:---:|:---:|:---:|:---:|:---:|:---:|
|文件|详细数据data|文件数据夹|读到文件内容|修改文件内容|执行文件内容|
|目录|文件名|可分类抽屉|读到文件名|修改文件名|进入该目录的权限|  

任何设备在Linux下面都是文件,概述文件种类:  
* 正规文件(regular file):第一个属性为\[-\]
    * 纯文本文件(ASCII)
    * 二进制文件(binary)
    * 数据格式文件(data)
* 目录(directory):第一个属性为\[d\]
* 链接文件(link):软连接,第一个属性为\[l\]
* 设备与设备文件(device):集中在 /dev 目录下
    * 区块(block)设备文件:存储数据,以提供系统随机存取的周边设备,即硬盘/软盘等,第一个属性为\[b\]
    * 字符(character)设备文件:键盘/鼠标等外设,第一个属性为\[c\]
* 数据接口文件(sockets):网络上的数据承接,第一个属性为\[s\],常在 /run或/tmp下
* 数据输送流文件(FIFO,pipe):FIFO是一种特殊的文件类型,主要目的在于解决多个程序同时存取一个文件所造成的错误问题.第一个属性为\[p\]

文件名需避免如下特殊字符:  
> ? > < ; & ! \[ \] | \ ' " ` ( ) { }  

Linux目录详解:  

|目录|应放置文件内容|
|:---|:---|
|FHS必需||
|/bin|可执行文件目录,通用级别,可被root和一般账号使用|
|/boot|开启所用文件,包括Linux核心文件以及开机菜单与开机所需配置文件等|
|/dev|任何设备和周边设备都以文件的形式存在于该目录|
|/etc|系统主要的配置文件保存于此|
|/lib|开机时会用到的函数库,以及在/bin或/sbin下指令所调用的函数库|
|/media|可移除的设备,光盘/软盘/DVD等|
|/mnt|类似/media,用以暂时挂载|
|/opt|第三方软件安装目录,之前习惯放置于/usr/local下|
|/run|系统开机后所产生的各项信息,可用内存来仿真|
|/sbin|只用root才能利用来进行系统配置的指令,<br>/sbin下的为开机过程中所需要的,里面包括开机/修复/还原系统所需的指令<br>/usr/sbin下的为某些服务器软件程序<br>/usr/local/sbin下为本机自行安装的软件所产生的系统可执行文件|
|/srv|一些网络服务启动之后,这些服务所需要取用的数据目录|
|/tmp|让一般使用者或者是正在执行的程序暂时放置文件的地方|
|/usr|...|
|/var|...|
|FHS可存||
|/home|系统默认的使用者主文件夹|
|/lib<qual>|存放与/lib不同格式的二进制函数库|
|/root|管理员主文件夹|
|others||
|/lost+found|使用标准的ext2/ext3/ext4文件系统格式才会产生的一个目录,<br>目的在于当文件系统发生错误时,	将一些遗失的片段放置到这个目录下|
|/proc|放置的数据都是在内存当中,<br>例如系统核心、行程信息(process)、周边设备的状态及网络状态等等。<br>因为这个目录下的数据都是在内存当中,<br>所以本身不占任何硬盘空间啊!<br>比较重要的文件例如:/proc/cpuinfo,/proc/dma,/proc/interrupts,/proc/ioports,/proc/net/*等等|
|/sys|跟/proc非常类似,也是一个虚拟的文件系统,<br>主要也是记录核心与系统硬件信息较相关的信息。|

![Linux目录架构](https://timgsa.baidu.com/timg?image&quality=80&size=b9999_10000&sec=1575962057&di=44c2cfb4a5a199c44fc7b6422794f035&imgtype=jpg&er=1&src=http%3A%2F%2Fimg2018.cnblogs.com%2Fblog%2F1411865%2F201810%2F1411865-20181017182551628-1317700407.png "目录架构")

## Linux文件和目录管理
执行指令时,系统会依照PATH的设置去每个PATH定义的目录下搜寻文件名为该指令的可执行文件.  
```bash
echo $PATH
# /home/lab/anaconda2/envs/python3.7/bin:/home/lab/anaconda2/condabin:/home/lab/bin:/home/lab/.local/bin:/usr/local/sbin:/usr/local/bin:/usr/sbin:/usr/bin:/sbin:/bin:/usr/games:/usr/local/games:/snap/bin:/usr/local/cuda/bin:/opt/jdk8/bin:/opt/jdk8/jre/bin:/opt/ant/bin:/opt/mvn/bin:/usr/local/go/bin
```  
关于PATH的应知应会:  
* 不同用户默认的PATH不同
* PATH是可以修改的
* 使用绝对路径或相对路径直接指定某个指令的文件名来执行,会比搜寻PATH来的正确
* 指令应该要放置到正确的目录下,执行才会比较方便
* 本目录(.)最好不要放到PATH当中

### 指令文件名的搜寻  
* which (寻找"可执行文件")
* whereis (由一些特定的目录中寻找文件/目录名)
* locate / updatedb
    locate [-ir] keyword  
    选项和参数:  
    * -i: 忽略大小写的差异
    * -c: 不输出,仅计算找到的文件数量
    * -l: 仅输出几行的意思,如输出五行 -l 5
    * -S: 输出数据库文件的相关信息
    * -r: 后接正则表达式的显示方式
* find  
    find [PATH] [option] [action]    
    选项和参数:    
    * 与时间有关的选项:共有-atime,-ctime与-mtime,以-mtime说明  
        * -mtime n:n为数字,意义为在n天之前的“一天之内”被更动过内容的文件;  
        * -mtime +n:列出在n天之前(不含n天本身)被更动过内容的文件文件名;  
        * -mtime -n:列出在n天之内(含n天本身)被更动过内容的文件文件名。  
        * -newer file:file为一个存在的文件,列出比file还要新的文件文件名  
        将过去系统上面24小时内有更动过内容(mtime)的文件列出:    
        find / -mtime 0    
        寻找/etc下面的文件,如果文件日期比/etc/passwd新就列出:    
        find /etc -newer /etc/passwd    
        示例:    
            * +4代表大于等于5天前的文件名:ex> find /var -mtime +4  
            * -4代表小于等于4天内的文件文件名:ex> find /var -mtime -4  
            * 4则是代表4-5那一天的文件文件名:ex> find /var -mtime 4    
    * 与使用者或群组名称有关的参数   
    * 与文件权限及名称有关的参数    
        * -name filename  
        * -size [+-]SIZE  
        * -type TYPE : 一般正规文件(f),设备文件(b,c),目录(d),链接文件(l)..  
        * -perm mode  
        * -perm -mode  
        * -perm /mode  
        示例:找出/home/lab/test文件夹中名字包含logo的文件夹    
        find /home/lab/test -type d -name "*logo*"  
    * 额外可进行的动作   
    示例:找出/etc下,文件大小介于50k到60k之间的文件,并把权限完整的列出  
    find /etc -size +50k -a -size -60k -exec ls -l {} \;  
    示例:找出/etc下,文件大小大于50k且文件所属人不是root的文件名,并将权限完整列出  
    find /etc -size +50k -a ! -user root -type f -exec -l {} \;  
    示例:找到的文件使用ls -l列出来    
    find /usr/bin /usr/sbin -perm /7000 -exec -l {} \;
        * {}代表的是“由find找到的内容”   
        * -exec一直到\;是关键字,代表find额外动作的开始(-exec)到结束(\;),在这中间的就是find指令内的额外动作  
        * 因为“;”在bash环境下是有特殊意义的,因此利用反斜线来跳脱  

## Linux磁盘与文件系统管理
操作系统分区完毕后为什么要进行格式化(format)?  
因为每个操作系统所设置的文件属性/权限并不相同,为了存放这些文件所需的数据,需要将分区格式化,以称为操作系统能够利用的"文件系统格式(filesystem)".  
### 文件系统的简单操作   
```
# df [-ahikHTm] [目录或文件名]
-a		:列出所有的文件系统,包括系统特有的	/proc	等文件系统;
-k		:以	KBytes	的容量显示各文件系统;
-m		:以	MBytes	的容量显示各文件系统;
-h		:以人们较易阅读的	GBytes,	MBytes,	KBytes	等格式自行显示;
-H		:以	M=1000K	取代	M=1024K	的进位方式;
-T		:连同该	partition	的	filesystem	名称	(例如	xfs)	也列出;
-i		:不用磁盘容量,而以	inode	的数量来显示
```  
```
# ln [-sf] source destination
-s      :缺省hard link, -s -- symbolic link
-f      :已存在则强制移除创建
```  
```
lsblk 列出系统上的所有磁盘列表
blkid 获取磁盘UUID TYPE等信息
# lsblk [-dfimpt] [device]
-d		:仅列出磁盘本身,并不会列出该磁盘的分区数据
-f		:同时列出该磁盘内的文件系统名称
-i		:使用	ASCII	的线段输出,不要使用复杂的编码	(再某些环境下很有用)

-t		:列出该磁盘设备的详细数据,包括磁盘伫列机制、预读写的数据量大小等
```  
